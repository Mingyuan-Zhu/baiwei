## Load data

```python
import pandas as pd
CP = pd.read_excel('(Replace)Commercial Planning.xlsx')
HSV = pd.read_excel('(Replace)Historical Sales Volume 2016.1-2019.11.xlsx')
PD = pd.read_excel('(Replace)Product Master Data.xlsx')
```

```python
HSV2 = HSV.groupby(['YM','SalesRegion','SKU Code'])['VolumeHL'].sum().to_frame("VolumeHL/M").reset_index()
HSV2
```

## Merge 

Add product data

```python
data1 = pd.merge(HSV2,PD,on='SKU Code',how='left',sort= False)
data1
```

```python
CP
```

Add Commercial Planning

```python
data = pd.merge(data1,CP, how='inner',on=['YM','SalesRegion','SKU Code'])
data
```

## 填补缺失值Price

```python
zuhe = []
for i,j in data.groupby(['SalesRegion','SKU Code'])['Price']:
    zuhe.append(i)
zuhe
```

```python
for region,sku in zuhe:
    missValue = data[(data['SalesRegion']== 'Heilongjiang') & (data['SKU Code']== 3) & (data['Price'].notnull())]['Price'].mean()

    a = data[(data['SalesRegion']== region) & (data['SKU Code']== sku) & (data['Price'].isnull())].index
    for i in a:
        data.loc[i,'Price'] = missValue
```

```python
data1 = data

#数值类型进行归一化(0-1范围)
def feature_standarize(data,columns):
    combine_col=[]
    for j in columns:
        min_=data[j].min()
        max_=data[j].max()
        standard_col=data[j].apply(lambda x:(x-min_)/(max_-min_))#标准化为0-1范围
        #standard_col=data[j].apply(lambda x:(x-data[j].min())/(data[j].max()-data[j].min()))这样运行很慢，因为每次都要找min，max
        combine_col.append(standard_col)
    data_new2=pd.concat(combine_col,axis=1)
    return data_new2

data_CompetitionDistance=feature_standarize(data1,['VolumeHL/M'])
data_CompetitionDistance  = data_CompetitionDistance.rename(columns = {'VolumeHL/M':'volume_t'})


data1 =pd.concat([data_CompetitionDistance ,data1],axis=1)
data1
```
